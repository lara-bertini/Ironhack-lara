Construa um jogo utilizando os conceitos de programação aprendidos até o momento. Não precisa ser complicado: diversos jogos, como forca ou jogo da velha, são simples de se implementar. Simples ou complexo, um jogo em Python em geral contém um loop do jogo (muitas vezes um loop ```while True```) com as etapas abaixo:

1. Entrada do Usuário;
1. Atualização do estado do jogo;
1. Atualização das informações aos jogadores (via ```print``` por exemplo).

Além dos jogos acima, podemos utilizar jogos de azar (roleta por exemplo), cujas regras em geral são simples, ou então jogos da Antiguidade, como os descritos no site https://www.ancientgames.org/games/

````python
import numpy as np

class Jogo_Velha:
    def __init__(self):
        self.tabuleiro = np.zeros((3,3), dtype=int)

    def print_tabuleiro(self):
        print('\n 1 | 2 | 3\t',self.tabuleiro[0][0],'|',self.tabuleiro[0][1],'|',self.tabuleiro[0][2],'\n','----------','\n','4 | 5 | 6\t',self.tabuleiro[1][0],'|',self.tabuleiro[1][1],'|',self.tabuleiro[1][2],'\n','----------','\n','7 | 8 | 9\t',self.tabuleiro[2][0],'|',self.tabuleiro[2][1],'|',self.tabuleiro[2][2],'\n')
        
    def check_ganhador(self):
        print(self.tabuleiro[0][0])
        
    def check_game_over(self):
        # retorna 1: game over
        # retorna 0: jogo não acabou
        if ((self.tabuleiro==0).sum())==0:
            return 1
        else:
            return 0

    def check_game_won(self):
        # retorna 1: jogador 1 venceu o jogo
        # retorna 2: jogador 2 venceu o jogo
        # retorna 0: jogo não acabou / ninguém venceu o jogo

        boolean_diagonal_1 = self.tabuleiro[0][0]==self.tabuleiro[1][1] and self.tabuleiro[1][1]==self.tabuleiro[2][2] and self.tabuleiro[2][2]!=0
        boolean_diagonal_2 = self.tabuleiro[0][2]==self.tabuleiro[1][1] and self.tabuleiro[1][1]==self.tabuleiro[2][0] and self.tabuleiro[2][0]!=0
        boolean_linha_1 = self.tabuleiro[0][0]==self.tabuleiro[0][1] and self.tabuleiro[0][1]==self.tabuleiro[0][2] and self.tabuleiro[0][2]!=0
        boolean_linha_2 = self.tabuleiro[1][0]==self.tabuleiro[1][1] and self.tabuleiro[1][1]==self.tabuleiro[1][2] and self.tabuleiro[1][2]!=0
        boolean_linha_3 = self.tabuleiro[2][0]==self.tabuleiro[2][1] and self.tabuleiro[2][1]==self.tabuleiro[2][2] and self.tabuleiro[2][2]!=0
        boolean_coluna_1 = self.tabuleiro[0][0]==self.tabuleiro[1][0] and self.tabuleiro[1][0]==self.tabuleiro[2][0] and self.tabuleiro[2][0]!=0
        boolean_coluna_2 = self.tabuleiro[0][1]==self.tabuleiro[1][1] and self.tabuleiro[1][1]==self.tabuleiro[2][1] and self.tabuleiro[2][1]!=0
        boolean_coluna_3 = self.tabuleiro[0][2]==self.tabuleiro[1][2] and self.tabuleiro[1][2]==self.tabuleiro[2][2] and self.tabuleiro[2][2]!=0

        
        if boolean_diagonal_1 or boolean_diagonal_2 or boolean_linha_1 or boolean_linha_2 or boolean_linha_3 or boolean_coluna_1 or boolean_coluna_2 or boolean_coluna_3:
            if boolean_diagonal_1:
                if self.tabuleiro[0][0]+self.tabuleiro[1][1]+self.tabuleiro[2][2]==3:
                    return 1
                else: return 2
            elif boolean_diagonal_2:
                if self.tabuleiro[0][2]+self.tabuleiro[1][1]+self.tabuleiro[2][0]==3:
                    return 1
                else: return 2
            elif boolean_linha_1:
                if self.tabuleiro[0][0]+self.tabuleiro[0][1]+self.tabuleiro[0][2]==3:
                    return 1
                else: return 2       
            elif boolean_linha_2:
                if self.tabuleiro[1][0]+self.tabuleiro[1][1]+self.tabuleiro[1][2]==3:
                    return 1
                else: return 2 
            elif boolean_linha_3:
                if self.tabuleiro[2][0]+self.tabuleiro[2][1]+self.tabuleiro[2][2]==3:
                    return 1
                else: return 2                
            elif boolean_coluna_1:
                if self.tabuleiro[0][0]+self.tabuleiro[1][0]+self.tabuleiro[2][0]==3:
                    return 1
                else: return 2                
            elif boolean_coluna_2:
                if self.tabuleiro[0][1]+self.tabuleiro[1][1]+self.tabuleiro[2][1]==3:
                    return 1
                else: return 2                
            elif boolean_coluna_3:
                if self.tabuleiro[0][2]+self.tabuleiro[1][2]+self.tabuleiro[2][2]==3:
                    return 1
                else: return 2 
        else: return 0

jogo1 = Jogo_Velha()      
i = False

while(jogo1.check_game_over()==0 and jogo1.check_game_won()==0):
    jogo1.print_tabuleiro()
    print('Jogador',(i+1),'Digite posição da jogada:')
    posicao = int(input())
    if posicao == 1:
        jogo1.tabuleiro[0][0] = int(i)+1
    elif posicao == 2:
        jogo1.tabuleiro[0][1] = int(i)+1
    elif posicao == 3:
        jogo1.tabuleiro[0][2] = int(i)+1
    elif posicao == 4:
        jogo1.tabuleiro[1][0] = int(i)+1
    elif posicao == 5:
        jogo1.tabuleiro[1][1] = int(i)+1
    elif posicao == 6:
        jogo1.tabuleiro[1][2] = int(i)+1
    elif posicao == 7:
        jogo1.tabuleiro[2][0] = int(i)+1
    elif posicao == 8:
        jogo1.tabuleiro[2][1] = int(i)+1    
    else: jogo1.tabuleiro[2][2] = int(i)+1 
    i = not i
    print('\n=========\n')
    
if(jogo1.check_game_won()!=0):
    print('jogador vencedor:', jogo1.check_game_won())
````
