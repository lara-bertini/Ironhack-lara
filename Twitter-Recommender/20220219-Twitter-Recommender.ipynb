```python
def get_name(user_id):
    endpoint = "https://api.twitter.com/2/users"
    query = {"ids" : user_id}
    response = requests.request('GET', url = endpoint, headers=header, params = query)
    results = response.json()
    return results['data'][0]['name']

def get_username(user_id):
    endpoint = "https://api.twitter.com/2/users"
    query = {"ids" : user_id}
    response = requests.request('GET', url = endpoint, headers=header, params = query)
    results = response.json()
    return results['data'][0]['username']
```

```python
def get_id(username):
    endpoint = "https://api.twitter.com/2/users/by/username/"+username
    response = requests.request('GET', url = endpoint, headers=header)
    results = response.json()
    return results['data']['id']
```

```python
# Pegando mais infos do usu√°rio
def get_user_all_infos(user_id):
    endpoint = "https://api.twitter.com/2/users/"+user_id
    query = {"user.fields" : "location", "user.fields" : "created_at", "user.fields" : "public_metrics"}
    response = requests.request('GET', url = endpoint, headers=header, params = query)
    results = response.json()
    return (results['data']['id'],
            results['data']['name'],
            results['data']['username'],
            results['data']['public_metrics']['followers_count'],
            results['data']['public_metrics']['following_count'],
            results['data']['public_metrics']['tweet_count'],
            results['data']['public_metrics']['listed_count'])
```

```python
def get_followers_number(user_id):
    endpoint = "https://api.twitter.com/2/users"
    query = {"ids" : user_id, "user.fields" : "public_metrics"}
    response = requests.request('GET', url = endpoint, headers=header, params = query)
    results = response.json()
    return results['data'][0]['public_metrics']['followers_count']
```

```python
# Returns a list of all users infos who are followers of the specified user ID
def get_followers(user_id):
    endpoint = "https://api.twitter.com/2/users/"+user_id+"/followers"
    response = requests.request('GET', url = endpoint, headers=header)
    results = response.json()
    try:
        followers_list = [get_user_all_infos(follower['id']) for follower in results['data']]
        return followers_list
    except KeyError:
        return None
    except TypeError:
        return None

followers_dataframe = get_followers(get_id("larabertini"))
print(followers_dataframe)
```

```python
# Get all the followers at a depth of 2 degrees of connection for a user ID, not counting celebrities
# https://pt.wikipedia.org/wiki/Teoria_dos_seis_graus_de_separa%C3%A7%C3%A3o

def two_degree_followers(user_id):
    users_list = list()
    ids_list = list()
    users_list.extend([get_user_all_infos(user_id)])
    num_followers_uid = get_followers_number(user_id)
    print(num_followers_uid)
    if 0<num_followers_uid<50:
        followers_all_infos = get_followers(user_id)
        print(followers_all_infos)
        if followers_all_infos!=None:
            users_list.extend(followers_all_infos)
            ids_list.extend([(user_id,f_id) for (f_id,f_name,f_username,f_followers_count,f_following_count,f_tweet_count,f_listed_count) in followers_all_infos])
            for first_degree_follower_id in followers_all_infos:
                aux_id = first_degree_follower_id[0]
                num_followers_auid = get_followers_number(aux_id)
                if 0 < num_followers_auid <50:
                    followers_aux_all_infos = get_followers(aux_id)
                    if followers_aux_all_infos!=None:
                        users_list.extend(followers_aux_all_infos)
                        ids_list.extend([(aux_id,f_id) for (f_id,f_name,f_username,f_followers_count,f_following_count,f_tweet_count,f_listed_count) in followers_aux_all_infos])
    return (users_list, ids_list)

two_tree_connections = two_degree_followers(get_id("larabertini"))
```

```python
# Get all the followers at a depth of 3 degrees of connection for a user ID, not counting celebrities
# https://pt.wikipedia.org/wiki/Teoria_dos_seis_graus_de_separa%C3%A7%C3%A3o

def three_degree_followers(user_id):
    users_list = list()
    ids_list = list()
    users_list.extend([get_user_all_infos(user_id)])
    num_followers_uid = get_followers_number(user_id)
    # seguidores diretos
    if 0<num_followers_uid<100:
        first_followers_all_infos = get_followers(user_id)
        if first_followers_all_infos!=None:
            users_list.extend(first_followers_all_infos)
            ids_list.extend([(user_id,f_id) for (f_id,f_name,f_username,f_followers_count,f_following_count,f_tweet_count,f_listed_count) in first_followers_all_infos])
            time.sleep(60*15)
            for first_degree_follower in first_followers_all_infos:
                first_degree_follower_id = first_degree_follower[0]
                num_followers_fist_degree = get_followers_number(first_degree_follower_id)
                # seguidores dos usu√°rios do meu primeiro n√≠vel de conex√£o
                if 0 < num_followers_fist_degree <100:
                    followers_first_degree_all_infos = get_followers(first_degree_follower_id)
                    if followers_first_degree_all_infos!=None:
                        users_list.extend(followers_first_degree_all_infos)
                        ids_list.extend([(first_degree_follower_id,f_id) for (f_id,f_name,f_username,f_followers_count,f_following_count,f_tweet_count,f_listed_count) in followers_first_degree_all_infos])
                        time.sleep(60*15)
                        for second_degree_follower in first_followers_all_infos:
                            second_degree_follower_id = second_degree_follower[0]
                            num_followers_second_degree = get_followers_number(second_degree_follower_id)
                            # seguidores dos usu√°rios do meu segundo n√≠vel de conex√£o
                            if 0 < num_followers_second_degree <100:
                                followers_second_degree_all_infos = get_followers(second_degree_follower_id)
                                if followers_second_degree_all_infos!=None:
                                    users_list.extend(followers_second_degree_all_infos)
                                    ids_list.extend([(second_degree_follower_id,f_id) for (f_id,f_name,f_username,f_followers_count,f_following_count,f_tweet_count,f_listed_count) in followers_second_degree_all_infos])
                                    time.sleep(60*15)
                                    for third_degree_follower in followers_second_degree_all_infos:
                                        third_degree_follower_id = third_degree_follower[0]
                                        num_followers_third_degree = get_followers_number(third_degree_follower_id)
                                        # seguidores dos usu√°rios do meu segundo n√≠vel de conex√£o
                                        if 0 < num_followers_third_degree <100:
                                            followers_third_degree_all_infos = get_followers(third_degree_follower_id)
                                            if followers_third_degree_all_infos!=None:
                                                users_list.extend(followers_third_degree_all_infos)
                                                ids_list.extend([(third_degree_follower_id,f_id) for (f_id,f_name,f_username,f_followers_count,f_following_count,f_tweet_count,f_listed_count) in followers_third_degree_all_infos])
                                                time.sleep(60*15)
    return (users_list, ids_list)

three_tree_connections = three_degree_followers(get_id("larabertini"))
```
