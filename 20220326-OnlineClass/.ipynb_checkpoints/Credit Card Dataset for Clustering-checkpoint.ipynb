```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
import seaborn as sns
```

```python
cc_ds = pd.read_csv('CC GENERAL.csv')
```

```python
cc_ds.head()
```

```python
cc_ds.info()
```

```python
cc_ds = cc_ds.drop(['CUST_ID'], axis='columns')
```

```python
cc_ds.head()
```

```python
cc_ds = cc_ds.dropna(axis='rows')
```

```python
cc_ds.info()
```

```python
norm = StandardScaler().fit(cc_ds)
cc_ds_norm = norm.transform(cc_ds)
```

```python
pca_cc = PCA()
pca_cc.fit_transform(pd.DataFrame(cc_ds_norm))
```

```python
pca_cc.explained_variance_
```

```python
plt.plot(pca_cc.explained_variance_ratio_)
```

```python
pca_cc_7 = PCA(7)
pca_cc_7 = pca_cc_7.fit_transform(pd.DataFrame(cc_ds_norm))
```

```python
col = ['PC_'+str(i) for i in range(1,8)]
```

```python
pca_cc_7_df = pd.DataFrame(pca_cc_7, columns = col)
type(pca_cc_7_df)
```

```python
pca_cc_7_df.head()
```

```python
sns.clustermap(pca_cc_7_df)
```

```python
sns.pairplot(pca_cc_7_df)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
Z = linkage(pca_cc_7_df, method='ward')
dendrogram_ = dendrogram(Z,p = 50, truncate_mode = 'lastp')
```

```python
from sklearn.cluster import AgglomerativeClustering
hierarchical = AgglomerativeClustering(n_clusters = 4, affinity='euclidean', linkage='ward')
hierarchical.fit(pca_cc_7_df)
pca_cc_7_df['hierarchy_predict'] = [str(x) for x in hierarchical.labels_]
```

```python
sns.pairplot(pca_cc_7_df, vars = ['PC_1', 'PC_2', 'PC_3', 'PC_4', 'PC_5', 'PC_6', 'PC_7'], hue = 'hierarchy_predict')
```

```python
pca_cc_7_df['hierarchy_predict'].value_counts()
```

```python
cc_ds['hierarchy_predict'] = pca_cc_7_df['hierarchy_predict']
```
